```python
import networkx
import math
import pickle
from collections import Counter
import matplotlib.pyplot as plt
from collections import Counter
G1_100 = networkx.read_gpickle("G1_100")
deg = [d for (n, d) in G1_100.degree()]
# print(deg)
# print(G1_100.number_of_nodes())
```

```python
list_all_x = []
list_all_y = []
```

```python
G_tmp = networkx.Graph()
def deg_dist(G_tmp):
    deg = [d for (n, d) in G_tmp.degree()]
    cnt_nodes = G_tmp.number_of_nodes()
    dict_deg_counts = Counter(deg)
    list_pairs = []
    for m in dict_deg_counts:
        pair = [m, dict_deg_counts[m]]
        list_pairs.append(pair)
    list_pairs.sort(key = lambda x:x[0])
#     print(list_pairs)
    x_cord = []
    y_cord = []
    for p in list_pairs:
        x_cord.append(math.log10(p[0]))
        y_cord.append(math.log10(p[1] / cnt_nodes))

    list_all_x.append(x_cord)
    list_all_y.append(y_cord)
#     print(x_cord)
#     print(y_cord)
```

```python
G1_100 = networkx.read_gpickle("G1_100")
G1_1000 = networkx.read_gpickle("G1_1000")
G1_10000 = networkx.read_gpickle("G1_10000")
deg_dist(G1_100)
deg_dist(G1_1000)
deg_dist(G1_10000)
```

```python
plt.style.use('seaborn-darkgrid')
plt.scatter(list_all_x[0], list_all_y[0], label= "For 10^2 nodes", color= "red",  marker= "o", s=40) 
plt.xlabel('log k')
plt.ylabel('log P(k)')
plt.legend() 
plt.show()
```

```python
G1_100 = networkx.read_gpickle("G1_100")
G1_1000 = networkx.read_gpickle("G1_1000")
G1_10000 = networkx.read_gpickle("G1_10000")
deg_dist(G1_100)
deg_dist(G1_1000)
deg_dist(G1_10000)
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(9,9))
plt.scatter(list_all_x[0], list_all_y[0], label= "For 10^2 nodes", color= "red",  marker= "o", s=40) 
plt.scatter(list_all_x[1], list_all_y[1], label= "For 10^3 nodes", color= "green",  marker= "o", s=40) 
plt.scatter(list_all_x[2], list_all_y[2], label= "For 10^4 nodes", color= "blue",  marker= "o", s=40) 
plt.xlabel('log k')
plt.ylabel('log P(k)')
plt.legend() 
plt.show()
```

```python
#Finding the values of x mean and y mean
x_cord = list_all_x[0]
y_cord = list_all_y[0]
mean_x = 0
mean_y = 0
ss_xy, ss_xx = 0, 0
for i in range(len(x_cord)):
    mean_x += x_cord[i]
    mean_y += y_cord[i]
    ss_xy = ss_xy + (y_cord[i] * x_cord[i])
    ss_xx = ss_xx + (x_cord[i] * x_cord[i])

mean_x = mean_x / len(x_cord)
mean_y = mean_y / len(y_cord)

ss_xy = ss_xy - (len(x_cord) * mean_x * mean_y)
ss_xx = ss_xx - (len(x_cord) * mean_x * mean_x)

slope = ss_xy / ss_xx
constant_val = mean_y - (slope * mean_x)
# print(f"Slope value is = {slope}")
# print(f"Constant value is = {constant_val}")
gamma_0 = -1 * slope
print(f"Gamma value for power law with 10^2 nodes = {gamma_0}")
```

```python
y_cord_pred_0 = []
for i in range(len(x_cord)):
    y_pred = (slope * x_cord[i]) + constant_val
    y_cord_pred_0.append(y_pred)
```
