```python
import pickle
import networkx
import math
import pickle
from collections import Counter
import matplotlib.pyplot as plt
from collections import Counter
Picklefile1 = open('degrees_growth', 'rb')
dict_growth = pickle.load(Picklefile1)
# print(dict_growth)
```

```python
x_cords = [[],[],[]]
y_cord_0 = dict_growth[100]
print(len(y_cord_0))
y_cord_1 = dict_growth[1000]
y_cord_2 = dict_growth[5000]
for ab in range(1, 10001):
    if ab >= 100 and ab < 1000:
        x_cords[0].append(ab)
    elif ab >= 1000 and ab < 5000:
        x_cords[0].append(ab)
        x_cords[1].append(ab)
    elif ab >= 5000 and ab <= 9997:
        x_cords[0].append(ab)
        x_cords[1].append(ab)
        x_cords[2].append(ab)

# print(x_cords[0])
# print(x_cords[1])
# print(x_cords[2])
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(9,9))
plt.xlim(0,10000)
# plt.ylim(0,10000)
plt.plot(x_cords[0], y_cord_0, label = "For node 100")
plt.plot(x_cords[1], y_cord_1, label = "For node 1000")
plt.plot(x_cords[2], y_cord_2, label = "For node 5000")
plt.xlabel('time')
plt.ylabel('degree')
plt.legend() 
plt.show()
```
